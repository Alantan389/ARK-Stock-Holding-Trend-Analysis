```python
### goal: find out the stocks share presentage for two different dataset (sep and oct), find out he trend of holding


import pandas as pd
import glob
path = r'/Users/balance/Desktop/OCT' # use your path
all_files = glob.glob(path + "/*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
pd.set_option('display.max_rows', None)
frame
```

```python
frame1=frame.rename(columns={'weight(%)':'weight_percentage'})
```

```python
frame1.weight_percentage.describe()
```

```python
bin_edges = [1,3,5,12.05]
bin_names = ['Not_Consider', 'Consider', 'Buy']

frame1['Holding']=pd.cut(frame1['weight_percentage'],bin_edges,labels=bin_names)
frame1
```

```python
Buy=frame1.query('Holding=="Buy"')
```

```python
Buy.company.value_counts()
```
